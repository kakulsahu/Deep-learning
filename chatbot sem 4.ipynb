```python
import numpy as np
import pandas as pd
import nltk

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import SimpleRNN
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import GRU

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
training_sentences = [
    "hello",
    "hi",
    "hey",
    "good morning",
    "bye",
    "see you",
    "thanks",
    "thank you",
    "how are you"
]

labels = [
    "greeting",
    "greeting",
    "greeting",
    "greeting",
    "goodbye",
    "goodbye",
    "thanks",
    "thanks",
    "status"
]
```

```python
from sklearn.preprocessing import LabelEncoder

tokenizer = Tokenizer()

tokenizer.fit_on_texts(training_sentences)

X = tokenizer.texts_to_sequences(training_sentences)

max_len = max(len(x) for x in X)

X = pad_sequences(X, maxlen=max_len, padding='post')

vocab_size = len(tokenizer.word_index) + 1

encoder = LabelEncoder()

y = encoder.fit_transform(labels)
```

```python
rnn_model = Sequential()

rnn_model.add(Embedding(vocab_size, 10, input_length=max_len))
rnn_model.add(SimpleRNN(64))
rnn_model.add(Dense(vocab_size, activation='softmax'))

rnn_model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

rnn_model.fit(X, y, epochs=300)
```

```python
lstm_model = Sequential()

lstm_model.add(Embedding(vocab_size, 10, input_length=max_len))
lstm_model.add(LSTM(64))
lstm_model.add(Dense(len(set(labels)), activation='softmax'))

lstm_model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

lstm_model.fit(X, y, epochs=300)
```

```python
gru_model = Sequential()

gru_model.add(Embedding(vocab_size, 10, input_length=max_len))
gru_model.add(GRU(64))
gru_model.add(Dense(vocab_size, activation='softmax'))

gru_model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

gru_model.fit(X, y, epochs=300)
```

```python
responses = {
    "greeting": "Hello!",
    "goodbye": "Goodbye!",
    "thanks": "Welcome!",
    "status": "I am fine."
}

def chatbot_response(model, text):

    sequence = tokenizer.texts_to_sequences([text])

    sequence = pad_sequences(sequence, maxlen=max_len)

    prediction = model.predict(sequence)

    predicted_class = encoder.inverse_transform(
        [np.argmax(prediction)]
    )[0]

    return responses[predicted_class]
```

```python
print(chatbot_response(rnn_model, "hello"))

print(chatbot_response(lstm_model, "hello"))

print(chatbot_response(gru_model, "hello"))
```

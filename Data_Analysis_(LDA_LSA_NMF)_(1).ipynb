#Data Preprocessing

```python
import nltk
import re
import string
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
# Remove URLs, citations [1], special characters
text = re.sub(r'http\S+', '', text)           # remove URLs
text = re.sub(r'\[\d+\]', '', text)           # remove citations
text = re.sub(r'[^a-zA-Z\s]', '', text)       # keep only alphabets
text = text.lower()                           # lowercase


# 3. Tokenization
tokens = word_tokenize(text)


# 4. Stopword Removal
stop_words = set(stopwords.words('english'))
tokens = [word for word in tokens if word not in stop_words]


# 5. Lemmatization
lemmatizer = WordNetLemmatizer()
tokens = [lemmatizer.lemmatize(word) for word in tokens]
```

```python
# 6. Final Clean Text
clean_text = " ".join(tokens)

print("\nCleaned Text Sample:\n", clean_text[:500])
```

#**LSA**(Latent Dirichlet Allocation)

```python
x=vectorizer.fit_transform([text])
```

```python
lsa=TruncatedSVD(n_components=3,random_state=42)
```

```python
lsa_matrix=lsa.fit_transform(x)
```

```python
terms=vectorizer.get_feature_names_out()

print('\nLSA Topics')
for i,comp in enumerate(lsa.components_):
  print(f"\nTopic {i+1}:")
  terms_weights=zip(terms,comp)
  sorted_terms=sorted(terms_weights,key=lambda x: x[1],reverse=True)
  for term,weight in sorted_terms[:6]:
    print(term)
```

#**NMF**(Non-negative Matrix Factorization)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
```

```python
vectorizer=TfidfVectorizer(stop_words='english')
X=vectorizer.fit_transform([text])
```

```python
nmf=NMF(n_components=3,random_state=42)
W=nmf.fit_transform(X)
H=nmf.components_
```

```python
print('\nNMF Topics')
feature_names=vectorizer.get_feature_names_out()
for topic_idx,topic in enumerate(H):
  print(f"\nTopic {topic_idx+1}:")
  top_words=[feature_names[i] for i in topic.argsort()[-6:]]
  print(top_words)
```
